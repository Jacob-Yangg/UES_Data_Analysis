# Cleaning Major data (replacing redundant responses)

```python


plot_data = surveydata
mask = plot_data.columns.str.contains(
    "Name|Year|UCSD Email|Do you intend on coming to UES's recruitment?|Any questions or comments?|Where did you hear about us?", regex=True)
columns_to_drop = plot_data.columns[mask]
plot_data = plot_data.drop(columns=columns_to_drop)


# Here, we are manually cleaning the Major column in plot_data. Since students were given the option to manually input their major in the survey, this created variability in the number of different majors when we went to group students by major, as
# the groupby() method would, for example, treat "Applied Math" and "Applied Mathematics" as two separate majors. Thus, we went through and reanmed all majors to fall under 1 of 9 categories, which you can find in the plot_data output. 
plot_data = plot_data.replace('Data science','Data Science', regex=True)
plot_data = plot_data.replace('international studies - international business ','International Business', regex=True)
plot_data = plot_data.replace('International businesses ','International Business', regex=True)
plot_data = plot_data.replace('Int bus','International Business', regex=True)
plot_data = plot_data.replace('International business','International Business', regex=True)
plot_data = plot_data.replace('International Studies - International Business','International Business', regex=True)

plot_data = plot_data.replace('International Studies - Economics','Economics', regex=True)

plot_data = plot_data.replace('Applied math','Applied Mathematics', regex=True)
plot_data = plot_data.replace('Applied math cs ','Applied Mathematics', regex=True)
plot_data = plot_data.replace('Applied Mathematics cs','Applied Mathematics', regex=True)

plot_data = plot_data.replace(['Cogsci ML', 'business psych', 'Environmental Systems', 'Data Science', 'PoliSci', 'Marine biology'],['Other', 'Other', 'Other', 'Other', 'Other', 'Other'],regex=True)
plot_data = plot_data.replace(['Computer Engineering', 'Mechanical engineering ', 'Computer Science '], ['Engineering', 'Engineering', 'Engineering'], regex=True)
plot_data = plot_data.replace(['MATH/CS', 'Applied Mathematics', 'Math & computer science', 'Math - cs', 'Statistics and probability', 'Probability and Statistics'], 
                              ['Applied Math / CS', 'Applied Math / CS','Applied Math / CS', 'Applied Math / CS', 'Applied Math / CS', 'Applied Math / CS'], regex=True) 
plot_data = plot_data.replace(['Economics and Communication Double major', 'Economics & Communication Double Major'], ['Economics', 'Economics'], regex=True)
plot_data = plot_data.replace(['undeclared', 'Undecided, thinking about Econ or Math-Econ'], ['Undeclared', 'Undeclared'], regex=True)

plot_data['Major'] = plot_data.Major.str.replace(r'(^.*Statistics.*$)', 'Math-Econ', regex=True)
plot_data['Major'] = plot_data['Major'].fillna('Other')


plot_data = plot_data.groupby(by=['Major']).sum()
plot_data.reset_index(inplace=True)
plot_data
```

# Visualizing Majors and Classes

```python
import matplotlib.pyplot as plt
import numpy as np
#plt.show()
plot_data.plot(x="Major", y=['1st', '2nd', '3rd', '4th'], kind = 'bar', figsize=(15,8))
plt.title("Attendees' Majors and Years")
plt.show()
```
